```python
import pandas as pd
data = pd.read_csv("CarPrice_assignment.csv").drop("car_ID",axis=1)
data.head()
```

```python
from pycaret.regression import *
st=setup(data=data,target="price",remove_multicollinearity=True,train_size=0.7
         ,polynomial_features=True,remove_outliers=True)
```

```python
bst=compare_models(include=["lr",'lasso','ridge','huber'])
```

```python
evaluate_model(bst)
```

```python
create_model(bst)
```

```python
tm=tune_model(bst)
```
