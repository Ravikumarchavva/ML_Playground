```python
from pycaret.classification import *
```

```python
import pandas as pd
train_data = pd.read_csv("train_featureEngineered.csv")
test_data = pd.read_csv("test_featureEngineered.csv")
data=pd.concat([train_data,test_data])
```

```python
st=setup(data=data.drop("customerID",axis=1),target="Churn",train_size=0.7,index=False,polynomial_degree=2,polynomial_features=True,fix_imbalance=True,
         feature_selection=True,n_features_to_select=0.9,profile=True)
```

```python
bst=compare_models(sort='recall')
```

```python
evaluate_model(bst)
```
